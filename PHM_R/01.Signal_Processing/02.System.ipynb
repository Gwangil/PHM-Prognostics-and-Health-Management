## 1. Linear Time-Invariant(LTI) Systems

<div id="systems" class="section level3">
<h3>1.1. Systems</h3>
A discrete-time system <span class="math inline">\(\mathcal{H}\)</span> is a transformation (a rule or formula) that maps a discrete-time input signal <span class="math inline">\(x\)</span> into a discrete-time output signal <span class="math inline">\(y\)</span>
<center>

</center>
</div>

<div id="linear-systems" class="section level3">
<h3>1.2. Linear Systems</h3>
<p>A system <span class="math inline">\(\mathcal{H}\)</span> is <strong>linear</strong> if it satisfies the following two properties:</p>
<strong>1) Scaling</strong> <span class="math display">\[\mathcal{H} \{ \alpha x \} = \alpha \mathcal{H}\quad \forall \alpha \in \mathbb{C} \]</span>
<center>

</center>
<strong>2) Additivity</strong> <span class="math display">\[\text{If}\ y_1 = \mathcal{H} \{ x_1 \}\ \text{and}\ y_2 = \mathcal{H} \{x_2\}\ \text{then}\ \mathcal{H}\{x_1 + x_2\}=y_1 + y_2 \]</span>
<center>

</center>
</div>
<div id="time-invariant-systems" class="section level3">
<h3>1.3. Time-Invariant Systems</h3>
A system <span class="math inline">\(\mathcal{H}\)</span> processing infinite-length signals is <strong>time-invariant</strong> (shift-invariant) if a time shift of the input signal creates a corresponding time shift in the output signal
<center>

</center>
</div>
<div id="linear-time-invariantlti-systems-1" class="section level3">
<h3>1.4. Linear Time-Invariant(LTI) Systems</h3>
<p>We usally consider <strong>Linear Time-Invariant(LST)</strong> systems.</p>

<h2>2. Convolution</h2>
<strong>Convolution</strong> is defined as the integral of the product of the two functions after one is reversed and shifted <span class="math display">\[y[n] = \sum_{m=-\infty}^\infty h[n-m]x[m] = x[n]*h[n]\]</span> output <span class="math inline">\(y[n]\)</span> came out by convolution of input <span class="math inline">\(x[n]\)</span> and system <span class="math inline">\(h[n]\)</span>

<div id="convolution-in-matrix-toeplitz-matrices" class="section level3">
<h3>2.1. Convolution in Matrix (= Toeplitz Matrices)</h3>
For infinite-Length Signals <span class="math display">\[y[n] = x[n] * h[n] = \sum_{m=-\infty}^\infty h[n-m]x[m] \\ = ... h[n+2]x[-2] + h[n+1]x[-1] + h[n]x[0] + h[n-1]x[1] + h[n-2]x[2] ...\]</span> It is inner product of <span class="math inline">\(h\)</span> vectors and <span class="math inline">\(x\)</span> <span class="math display">\[\begin{align*}
y[n] &amp;= \begin{bmatrix}
\cdots &amp; \bracevert &amp; \bracevert &amp; \bracevert &amp; \bracevert &amp; \bracevert &amp; \cdots\\
\cdots &amp; h[n+2] &amp; h[n+1] &amp; h[n] &amp; h[n-1] &amp; h[n-2]\ &amp; \cdots\\
\cdots &amp; \bracevert &amp; \bracevert &amp; \bracevert &amp; \bracevert &amp; \bracevert &amp; \cdots
\end{bmatrix}
\begin{bmatrix}
\vdots\\
x[-2]\\
x[-1]\\
x[0]\\
x[1]\\
x[2]\\
\vdots
\end{bmatrix}\\ \\ &amp;= H x
\end{align*}\]</span> Convolution is product of <span class="math inline">\(H\)</span> and <span class="math inline">\(x\)</span>
<center>

</center>
<p><span class="math display">\[h = h[0]\]</span></p>
</div>
<div id="impulse-response" class="section level3">
<h3>2.2. Impulse Response</h3>
(impulse): <span class="math inline">\(\delta[n] = \begin{cases} 1 \quad n = 0\\ 0 \quad \text{otherwise}\\ \end{cases}\)</span>
<center>

</center>
<ul>
<li><p>Output of system and <strong>delta function</strong> (impulse) is <span class="math inline">\(h\)</span>. So, We call <span class="math inline">\(h\)</span> the impulse response of the system</p></li>
<li>From <span class="math inline">\(h\)</span>, we can build matrix <span class="math inline">\(H\)</span>
<ul>
<li>Coulmns/row of <span class="math inline">\(H\)</span> are circularly shifted versions of the 0-th column/row</li>
<li><span class="math inline">\(h\)</span> contains all the information of the LTI system</li>
</ul></li>
<li>LTI systems are Toeplitz matrices(infinite-length signals)
<ul>
<li>Entries on the matrix diagonals are the same</li>
</ul></li>
</ul>

<div id="convolution-examples" class="section level2">
<h2>3. Convolution Examples</h2>
<p>Convolution = filtering in time domain</p>
<div id="convolution-demo" class="section level3">
<h3>3.1. Convolution Demo</h3>

```r
# Convolve a rectangular pulse with itself

# pulse
x <- c(0, 0, 0, 1, 1, 1, 0, 0, 0)

# convolve pulse with itself
conv <- function(x, y, conj = TRUE, type = c("circular", "open", "filter")){
  convolve(x, y[length(y):1], conj = conj, type = type)
}
y <- conv(x, x, type = 'open')

n <- 0:(length(y) - 1)

options(repr.plot.width=6, repr.plot.height=4)
par(mfrow = c(2, 1), mar = c(3, 2, 2, 1))
plot(x, col = 'deepskyblue', pch = 16, cex = 1, xlim = c(0, 16), ylim = c(-0.2, 1.2), main = 'pulse')
arrows(x0 = 1:9, y0 = 0, x1 = 1:9, y1 = x, col = 'deepskyblue', length = 0, angle = 0, lwd = 1)
abline(h = 0)

plot(n, y, pch = 16, col = 'deepskyblue', cex = 1, xlim = c(0, 16), ylim = c(-0.2, 3.2), main = 'pulse * pulse')
arrows(x0 = n, y0 = 0, x1 = n, y1 = y, col = 'deepskyblue', length = 0, angle = 0, lwd = 1)
abline(h = 0)
```

### 3.2. Denoising a Piecewise Smooth Signal

 * moving average(MA) filter
 * low-pass filter in time domain

```r
# piecewise smooth signal
N <- 50
n <- 0:(N - 1)

x <- sin(pi/N*n)*c(rep(1, N/2), rep(-1, N/2))
xn <- x + 0.1 * rnorm(N, 0, 1)

# construct moving average filter impulse response of length M
M <- 5
h <- rep(0, N)
h[1:5] <- 1 / M

# convolve noisy signal with impulse response
y <- conv(x, h, type = 'open')

par(mfrow = c(2, 2), mar = c(2, 2, 2, 1))
plot(x, pch = "", ylim = c(-1.5, 1.5), xlim = c(1, N), main = "piecewise smooth signal")
arrows(x0 = n, y0 = 0, x1 = n, y1 = x, col = 'deepskyblue', length = 0, angle = 0, lwd = 1)
abline(h = 0)

plot(xn, pch = "", ylim = c(-1.5, 1.5), xlim = c(1, N), main = "piecewise smooth signal + noise")
arrows(x0 = n, y0 = 0, x1 = n, y1 = xn, col = 'deepskyblue', length = 0, angle = 0, lwd = 1)
abline(h = 0)

plot(h, pch = "", ylim = c(-1.5, 1.5), xlim = c(1, N), main = "impulse response")
arrows(x0 = n, y0 = 0, x1 = n, y1 = h, col = 'deepskyblue', length = 0, angle = 0, lwd = 1)
abline(h = 0)

plot(y, pch = "", ylim = c(-1.5, 1.5), xlim = c(1, N), main = "convouluted output")
arrows(x0 = n, y0 = 0, x1 = n, y1 = y, col = 'deepskyblue', length = 0, angle = 0, lwd = 1)
abline(h = 0)
```

### 3.3. Edge Detection

```r
# haar wavelet edge detector
M <- 4
h <- rep(0, N)
h[1:2] <- -1 / M
h[3:4] <- 1 / M

# convolve noisy signal with impulse response
y <- conv(xn, h, type = 'open')

par(mfrow = c(2, 2), mar = c(2, 2, 2, 1))
plot(x, pch = "", ylim = c(-1.5, 1.5), xlim = c(1, N), main = "piecewise smooth signal")
arrows(x0 = n, y0 = 0, x1 = n, y1 = x, col = 'deepskyblue', length = 0, angle = 0, lwd = 1)
abline(h = 0)

plot(xn, pch = "", ylim = c(-1.5, 1.5), xlim = c(1, N), main = "piecewise smooth signal + noise")
arrows(x0 = n, y0 = 0, x1 = n, y1 = xn, col = 'deepskyblue', length = 0, angle = 0, lwd = 1)
abline(h = 0)

plot(h, pch = "", ylim = c(-1.5, 1.5), xlim = c(1, N), main = "impulse response")
arrows(x0 = n, y0 = 0, x1 = n, y1 = h, col = 'deepskyblue', length = 0, angle = 0, lwd = 1)
abline(h = 0)

plot(y, pch = "", ylim = c(-1.5, 1.5), xlim = c(1, N), main = "convouluted output")
arrows(x0 = n, y0 = 0, x1 = n, y1 = y, col = 'deepskyblue', length = 0, angle = 0, lwd = 1)
abline(h = 0)
```

### 3.4. Convolution on Audio

```r
x <- suppressWarnings(tuneR::readWave('../data_files/violin_origional.wav'))
Fs <- x@samp.rate
x <- x@left

x <- x / max(x)  # normalize
sound <- function(data, samp.rate) {
  W <- tuneR::normalize(tuneR::Wave(data, samp.rate = samp.rate, bit = 16), unit = "16")
  tuneR::play(W)
}
sound(x, Fs)  # play a wave file with sampling rate Fs

# plot wave file
plot(1:length(x)/Fs, x, type = 'l', lwd = 1, col = 'deepskyblue', xlab = 'time in sec', ylab = '')
```

```r
# impulse response in a closed room (by gunshot)
h <- suppressWarnings(tuneR::readWave('../data_files/gunshot.wav'))
Fs <- h@samp.rate
h <- h@left
h <- h / max(h)

# sound(h, Fs)

plot(1:length(h)/Fs, h, type = 'l', lwd = 1, col = 'deepskyblue', xlab = 'time in sec', ylab = '')
```

```r
y <- conv(x, h, type = 'open')
y <- y / max(y)

par(mfrow = c(2, 1), mar = c(2, 2, 2, 1))
plot(1:length(x)/Fs, x, type = 'l', lwd = 1, col = 'deepskyblue', main = 'original', xlab = '', ylab = '', xlim = c(0, 6))
plot(1:length(y)/Fs, y, type = 'l', lwd = 1, col = 'deepskyblue', main = 'convoluted', xlab = '', ylab = '', xlim = c(0, 6))

# image how the music played in a closed room sounds like

sound(y, Fs)
```

## 4. Non-linear System: Median Filter

$$y[n] = \text{median} \{x[n-k],\ ...,\ x[n+k]\}$$

 * There are nonlinear neighborhood operations that can be performed for the purpose of noise reduction that can do a better job of preserving edges than simple smoothing filters.
 * Median filters can do an excellent job of rejecting certain types of noise, in particular, "shot" or impulse noise (outlier in a time series) in which some individual pixels or signals have extreme values.
 
### 4.1. Removing Shot Noise in Audio

```r
x <- suppressWarnings(tuneR::readWave('../data_files/violin_origional.wav'))
Fs <- x@samp.rate
x <- x@left

x <- x / max(x)  # normalize
sound(x, Fs)  # play a wave file with sampling rate Fs
```

```r
# generate an audio signal with a salt and pepper noise: "on and off" pixels
shot_noise <- rep(0, length(x))
shot_noise[sample(1:length(x), length(x) * 0.04)] <- 1
x_noise <- x + shot_noise - mean(shot_noise);

sound(x_noise,Fs)
```

```r
# apply a linear low-pass filter

h = c(1,1,1)/3
x_avg = conv(x_noise, h, type = 'open')
sound(x_avg,Fs)  # does not work very well
```

```r
# apply a nonlinear filter
x_median = zoo::rollmedian(x_noise,7)
sound(x_median,Fs)  # WOW !!!
```

```r
par(mfrow = c(4, 1), mar = c(2, 2, 1, 1))
plot(x, type = 'l', col = 'deepskyblue', lwd = 1)
plot(x_noise, type = 'l', col = 'deepskyblue', lwd = 1)
plot(x_avg, type = 'l', col = 'deepskyblue', lwd = 1)
plot(x_median, type = 'l', col = 'deepskyblue', lwd = 1)
```
